# Model Training

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn import tree
```

# Feature and target segregation

```python
# dropping Loan ID
df.drop('LoanID', axis = 1 , inplace = True)
```

```python
df.head()
```

```python
x = df.drop('Default', axis = 1)
y = df['Default']
```

# Label Encoding

```python
# filtering only non numeric col
df_obj = df.select_dtypes(include=['object'])
```

```python
enc = LabelEncoder()
for col in df_obj:
    x[col] = enc.fit_transform(x[col])
```

Categorical(Textual) columns are now converted into Numerical data

```python
x.head()
```

# Standardization

```python
scaler = StandardScaler()
```

```python
df[df.columns]
```

```python
x[x.columns] = scaler.fit_transform(x[x.columns])
```

```python
x.head()
```

**Features are now standardized**

# Splitting the Data into train and test part

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape, x_test.shape
```

# Fitting the training Data

```python
dt = DecisionTreeClassifier(random_state=1, criterion='entropy')
```

```python
dt.fit(x_train,y_train)
```

```python
fig = plt.figure(figsize=(25,20))
tree.plot_tree(dt, filled = True, rounded = True, max_depth=3)
plt.show()
```

# Predictions

```python
y_pred = dt.predict(x_test)
```

# Accuracy Evaluation

```python
print('Accuracy:',accuracy_score(y_test,y_pred))
print('confusion_matrix:',confusion_matrix(y_test,y_pred))
print('Classification_report:',classification_report(y_test,y_pred))
```
